```python
f=open('chronic_kidney_disease.arff','r')
lines = f.readlines()
f.close()


old=(' yes',' no',',,',',ckd,')
new=('yes','no',',',',ckd')

fw=open('chronic_kidney_disease_new.arff', 'w')
for line in lines:
    for check, rep in zip(old, new):
        line = line.replace(check, rep)
  
    fw.write(line)
fw.close()
```

```python
import pandas as pd

with open('chronic_kidney_disease_new.arff', 'r') as fp:
    file_content = fp.readlines()

columns = []
len_attr = len('@attribute')

for line in file_content:
    if line.startswith('@attribute'):
        col_name = line[len_attr:].split()[0]
        col_name=col_name.replace("'","")
        columns.append(col_name)

data = []
with open('chronic_kidney_disease_new.arff', 'r') as f:
     for line in f.readlines()[30:]:
        line = line.replace('\n', '')
        data.append(line.split(','))

    
df = pd.DataFrame(data, columns=columns)
df = df[:-1]
df
```

```python
df.isnull().sum()
```

```python
import numpy as np

df[:]=df[:].replace('?',np.nan)
df[:]=df[:].replace('	?',np.nan)
```

```python
df[num_cols] = df[num_cols].apply(pd.to_numeric)
```
